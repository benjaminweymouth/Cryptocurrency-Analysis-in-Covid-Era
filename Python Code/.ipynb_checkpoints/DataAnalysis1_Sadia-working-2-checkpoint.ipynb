## Pre-Covid Analysis

Call each crypto currency data and Gold historical data and clean the data carrying out the following steps:

3. Drop unwanted columns
4. Rename column name to ticker

```python
# Starting with Bitcoin, we will call the historical data (pre-covid era)
btc_data = btc.history(start="2018-03-01", end="2021-06-01")
btc_data
```

```python
# Check the data types
btc_data.dtypes
```

```python
# Check for and drop any null values
btc_data.isnull().sum()
```

```python
# Drop unrequired columns, keeping only the closing values, and also renaming to match the ticker name
btc_close = btc_data.drop(columns=['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits']).rename(columns={"Close": "BTC"})
btc_close
```

```python
# Drop unrequired columns, keeping only the closing values, and also renaming to match the ticker name
eth_close = eth_data.drop(columns=['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits']).rename(columns={"Close": "ETH"})
eth_close
```

```python
# Lite coins historical data 
ltc_data = lite.history(start="2018-03-01", end="2021-06-01")
ltc_data
```

```python
# Gold historical data
gold_data = gold.history(start="2018-03-01", end="2021-06-01")
gold_data
```

```python
# Check for data types
gold_data.dtypes
```

```python
# Check for and drop any null values
gold_data.isnull().sum()
```

```python
# Drop unrequired columns, keeping only the closing values, and also renaming to match the ticker name
gold_close = gold_data.drop(columns=['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits']).rename(columns={"Close": "Gold"})
gold_close
```

```python
# Join the 4 data frames 
# Only compare dates where return data exists for all the stocks (drop NaNs)
combined_df = pd.concat([btc_close, eth_close, ltc_close, gold_close], axis = 1, join = "inner").dropna().copy()
combined_df.head()
```

### Quantitative Analysis

To conduct a quantitative analysis we will calculate and plot the following for all portfolios:
1. Daily returns
2. Cumulative returns

```python
# Calculation for daily returns:
daily_returns = combined_df.pct_change().dropna().copy()
daily_returns.head()
```

```python
# Calculation for cumulative returns
cumulative_returns = (1 + daily_returns).cumprod() - 1
cumulative_returns.head()
```

```python
# Read in the data for the plot
daily_returns.plot(title="Daily Returns", figsize=(12,7))
cumulative_returns.plot(title="Cumulative Returns", figsize=(12,7))
```

### Risk Analysis

Now, to determine the _risk_ of each portfolio:

1. We will create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the Gold.
5. Calculate the Annualized Standard Deviation.

```python
# Box plot to visually show risk
boxplot = daily_returns.boxplot(grid=False, rot=45, fontsize=10, figsize=(10,5))
```

```python
# Calculate the daily standard deviations of all portfolios
volatility = daily_returns.std()
volatility
```

```python
# Calculate the daily standard deviation of Gold
volatility_gold = daily_returns['Gold'].std()

# Determine which portfolios are riskier than Gold
daily_returns.std()>volatility_gold
```

```python
# Calculate the annualized standard deviation (252 trading days)
volatility = daily_returns.std() * np.sqrt(252)
volatility
```

### Calculating and plotting the correlation

```python
# Calculate the correlation
correlation = daily_returns.corr()

# Display the correlation matrix
correlation.plot(figsize=(10,5), title="Correlation")
correlation.style.background_gradient(cmap="Pastel1")
```
